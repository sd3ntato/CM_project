# Visualization of qadratic functions

```python
import numpy as np
import plotly.graph_objects as go
```

```python
Q = np.array([
              [1,0],
              [0,1]
])
q = np.array([-1,2]).reshape(-1,1)
print(Q)
print()
print(q)

# pointwise computation of simple quadratic functions
def simple_quadratic(X,Y):
  s = np.size(X) # lenght of data
  z = np.zeros((s,s)) # allocate matrix of points for surface plot
  for i,x1 in enumerate(X):
    for j,x2 in enumerate(Y):
      x = np.array([x1,x2]).reshape(2,1) # wanna put point in a column vector to compute function value in vectorial notation
      z[i,j] = ( (x.T @ Q @ x) + (q.T @ x) )[0,0] + (2*np.random.rand()-1) # computation of function value with stard formula
  return z
```

```python
# definition of arguments
x, y = np.linspace(-5, 5, 100), np.linspace(-5, 5, 100)

# production of function data
z = simple_quadratic(x,y)

#actual plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout()
fig.show()
```
